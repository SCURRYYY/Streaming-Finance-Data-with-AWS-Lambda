```python
import pandas as pd
import re
import datetime
import matplotlib.pyplot as plt
df=pd.read_csv('results.csv')
df.head()
```

## Graph 1 : The Stock Price by Stocks 

The bar chart shows that SHOP has the highest price. 

```python
#the bar plot for the max_high by hours 
gp1 = df.groupby("name")['max_high'].mean().plot(kind='bar',title='The Maximum Price by Stocks')
gp1
```

```python
#the scatterplot for the max_high by hours 
plt.scatter('hour', 'max_high', data = df )
plt.title('The Maximum Stock Price by Hours')
plt.xlabel('Hour')
plt.ylabel('Max Stock Price')
```

## Graph 2 : The Stock Price by Hours

The max_high is the highest ar 10am.

```python
gp2 = df.groupby("hour")['max_high'].mean().plot(kind='bar',title='The Maximum Stock Price by Hours')
gp2
```

## Graph 3 : How the Highest Stock Price Changes by Hours for Each Stock on on 14th May, 2020?

```python
gp5 = df.groupby('name').plot(x='hour', y='max_high',title='Change of Stock Price by Hour')
gp5
```
